```python
import pandas as pd

# Step 1: Create a DataFrame with more data
data = {
    'Name': ['Alice', 'Bob', 'Cathy', 'David', 'Eva'],
    'Age': [29, 35, 28, 40, 22],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Boston']
}
df = pd.DataFrame(data)

# Step 2: Basic operations on DataFrame
print("Initial DataFrame:")
print(df)

# View first and last rows
print("\nFirst 3 rows:")
print(df.head(3))  # First 3 rows

print("\nLast 2 rows:")
print(df.tail(2))  # Last 2 rows

# DataFrame summary
print("\nDataFrame Info:")
print(df.info())

# Step 3: Selecting and Filtering
print("\nSelect 'Name' column:")
print(df['Name'])

print("\nSelect multiple columns ('Name' and 'City'):")
print(df[['Name', 'City']])

print("\nFilter rows where Age > 30:")
print(df[df['Age'] > 30])

# Step 4: Adding, Modifying, and Dropping Columns
df['Salary'] = [70000, 85000, 62000, 90000, 55000]  # Add new column
print("\nAfter adding 'Salary' column:")
print(df)

df['Age'] = df['Age'] + 1  # Increment age by 1
print("\nAfter incrementing 'Age':")
print(df)

df = df.drop('Salary', axis=1)  # Drop the Salary column
print("\nAfter dropping 'Salary' column:")
print(df)

# Step 5: Sorting and Handling Missing Values
print("\nSorted by Age:")
print(df.sort_values(by='Age'))

df.loc[5] = ['Frank', None, 'Seattle']  # Add a row with missing Age
print("\nAfter adding a row with missing values:")
print(df)

df['Age'] = df['Age'].fillna(0)  # Fill missing Age with 0
print("\nAfter filling missing Age with 0:")
print(df)

# Step 6: Grouping and Aggregation
print("\nAverage Age by City:")
print(df.groupby('City')['Age'].mean())

print("\nAggregate stats on Age:")
print(df.agg({'Age': ['mean', 'max']}))

# Step 7: Merging and Concatenating DataFrames
data2 = {
    'City': ['New York', 'Chicago', 'Seattle'],
    'Population': [8500000, 2700000, 750000]
}
df2 = pd.DataFrame(data2)

merged_df = pd.merge(df, df2, on='City', how='left')  # Merge on 'City'
print("\nMerged DataFrame with Population data:")
print(merged_df)

df_part1 = df.iloc[:3]
df_part2 = df.iloc[3:]
concatenated = pd.concat([df_part1, df_part2], axis=0)  # Concatenate rows
print("\nConcatenated DataFrame:")
print(concatenated)

# Step 8: Using Apply and Pivot Tables
df['Age Category'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Adult')
print("\nAfter adding 'Age Category':")
print(df)

pivot = df.pivot_table(values='Age', index='City', aggfunc='mean')
print("\nPivot table (Average Age by City):")
print(pivot)

# Step 9: Working with DateTime
df['Joining Date'] = pd.to_datetime(['2023-01-01', '2023-02-15', '2023-03-10', '2023-04-01', '2023-05-20', '2023-06-30'])
print("\nAfter adding 'Joining Date':")
print(df)

print("\nExtract Year from 'Joining Date':")
print(df['Joining Date'].dt.year)

# Step 10: Handling Duplicates and Renaming Columns
print("\nCheck for duplicates:")
print(df.duplicated())

df = df.drop_duplicates()  # Remove duplicates
print("\nAfter removing duplicates:")
print(df)

df = df.rename(columns={'Name': 'Employee Name', 'City': 'Location'})
print("\nAfter renaming columns:")
print(df)

# Step 11: Exporting Data
df.to_csv('output.csv', index=False)  # Save as CSV
df.to_excel('output.xlsx', index=False)  # Save as Excel
print("\nData exported to 'output.csv' and 'output.xlsx'")
```
